## Analysis of Food Production.

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
```

```python
dfp = pd.read_csv('/content/Africa Food Production (2004 - 2013).csv') 
```

Let's find out a few things about our first dataset

```python
dfp.head(25)
```

```python
dfp.shape
```

There are four columns and 23110 rows in the dataset wow!!, Don't worry, with few lines of codes, we can extract the most important information from the dataset.

```python
dfp.nunique()
```

Told yaa!! There are 45 different African countries documented in this dataset over a period of 10 years. Interesting.

```python
dfp['Country'].unique()
```

And these are the unique countries.

```python
cvp = dfp.groupby('Country')['Value'].mean().sort_values(ascending=False)
cvp.plot(kind='bar',title='Average Production Per Country', color='green', figsize=[15,5]);
```

From this dataset, Nigeria has been the highest producing African country in terms of the quatity of food items produced for ten years(2004-2013). Nigeria my country!

```python
cvp_t10 = dfp.groupby('Country')['Value'].mean().nlargest(10).sort_values(ascending=False)
cvp_t10.plot(kind='bar',title='Top Ten Most Producing Country (By Quantity)', color='green', figsize=[15,5]);
```

```python
cvp_b10 = dfp.groupby('Country')['Value'].sum().nsmallest(10).sort_values(ascending=False)
cvp_b10.plot(kind='bar',title='Ten Least Producing Country (By Quantity)', color='green', figsize=[15,5]);
```

```python
cip = dfp.groupby('Country')['Item'].nunique().sort_values(ascending=False)
cip.plot(kind='bar',title='No. of Distinct Items Produced Per Country', color='green', figsize=[15,5]);
```

```python
cip_t10 = dfp.groupby('Country')['Item'].nunique().nlargest(10).sort_values(ascending=False)
cip_t10.plot(kind='bar',title='Top Ten Most Producing Country (By Variety)', color='green', figsize=[15,5]);
```

```python
dfp.groupby('Country')['Item'].nunique().sort_values(ascending=False)
```

Above is a list of countries and the total number of various food items that they produce.

```python
cip_b10 = dfp.groupby('Country')['Item'].nunique().nsmallest(10).sort_values(ascending=False)
cip_b10.plot(kind='bar',title='Ten Least Producing Country (By Variety)', color='green', figsize=[15,5]);
```

```python
cip_t10
```

>
>Countries like Egypt, South Africa, United Republic of Tanzania, Kenya and Morocco are in the top ten producing countries by quantity and variety.

>Comparing the bottom ten producing countries in terms of the quatity of items produced with the bottom ten producing countries in terms of the varieties of items produced, we can see that most of the countries that produce the least quatity of food items produce fewer varieties, hence, we can infer that producing wider range of food items can increase the producing power of a country. 
>
>Countries like Gabn, Mauritania, Botswana, Gambia, Lesotho, Sao Tome and Principe, Cabo Verde and Djibouti are in the bottom ten producing countries by quantity and variety.

```python
dfp['Item'].unique()
```

```python
ivp = dfp.groupby('Item')['Value'].mean().sort_values(ascending=False)
ivp.plot(kind='bar',title='Produced Items', color='orange', figsize=[15,5]);
```

>These are the various food items produced in the African countries found in the dataset. For easier understanding, these items have been arranged in decreasing order of frequency of production in the 45 African countries. 
>
>We can see from the above bar chart that most countries produce 'Cassava and Products' and the fewest number of countries produce 'Aquatic Animals, others'.

```python
dfp.groupby('Item')['Country'].nunique().sort_values(ascending=True)
```

```python

ivp_t20 = dfp.groupby('Item')['Value'].mean().nlargest(20).sort_values(ascending=False)
ivp_t20.plot(kind='bar',title='Top Twenty Most Produced Item', color='orange', figsize=[15,5]);
```

```python
ivp_b20 = dfp.groupby('Item')['Value'].mean().nsmallest(20).sort_values(ascending=False)
ivp_b20.plot(kind='bar',title='Twenty Least Produced Item', color='orange', figsize=[15,5]);
```

>From the year 2004 to 2013 average production has been on the rise and this can be linked with the increasing population in African countries. 
>
>From the graph below, there was a drop in average production from 2006 to 2007 but it picked up again in subsequent years.

```python
yvp = dfp.groupby('Year')['Value'].mean()
yvp.plot(kind='line', figsize=[15,5], title='Average production Per Year', xticks=dfp['Year']);
```

We can see from above that the average production of food increased over the years. It is also worthy of note that there was a drop in food production from 2006 to 2007 but the production increased in subsequent years.

```python
yvp = dfp.groupby('Year')['Value'].median()
yvp.plot(kind='line', figsize=[15,5], title='Median production Per Year', xticks=dfp['Year']);
```

The median production gives a more appreciable trend in the food production over the years. There is still an increment in food production but in a steeper manner.

```python
yvp = dfp.groupby('Year')['Value'].sum()
yvp.plot(kind='bar',title='Total Production Per Year', figsize=[15,5]);
```

```python
sns.boxplot(x='Year', y='Value', data= dfp)
plt.yscale('log')
plt.title('Box plot of Food Production By Year');
```

From the above box plot, we can infer that there was no lower outlier but majorly higher outliers. For better visualisations, the box plot was done using the logarithmic scale due to the large values.

## Analysis of Food Consumption.

```python
dfs = pd.read_csv('/content/Africa Food Supply (2004 - 2013).csv')
```

```python
dfs.head(20)
```

```python
dfs.shape
```

```python
dfs.nunique()
```

```python
cvs = dfs.groupby('Country')['Value'].mean().sort_values(ascending=False)
cvs.plot(kind='bar',title='Average Consumption Per Country', color='red', figsize=[15,5]);
```

Egypt is the highest consuming country over the ten years period (2004 - 2013). Well, that makes sense because they are also one of the top ten countries that produce large quantity of food items. This will cater for their high consumption rate.
Guess what doesn't make sense? Djibouti which ranked bottom in terms of quantity and variety of food items produced is the 22nd most consuming African country over these years. The country most have resorted to importing food items to cater for the consumption rate of their populace.

```python
yvs = dfp.groupby('Year')['Value'].mean()
yvs.plot(kind='line', figsize=[15,5], title='Average Consumption Per Year', color='red', xticks=dfp['Year']);
```

```python
yvs = dfp.groupby('Year')['Value'].median()
yvs.plot(kind='line', figsize=[15,5], title='Median Consumption Per Year', color='red', xticks=dfp['Year']);
```

```python
sns.boxplot(x='Year', y='Value', data= dfs)
plt.title('Box plot of Food Production By Year');
```

This box plot accomodates more range than the one we've seen earlier. From the box plot of the food consumption dataset, we can see the upper and lower outliers, the median values, interquartile ranges, upper and lower quartiles. 
We can also observe that the upper outliers are more than the lower outliers in terms of distance from the normal range.
